## Import modules

```python
import numpy as np
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate and plot random data

```python
x = 20 #20 random points
y = np.random.random(x)
```

```python
for i in range(len(y)):
    y[i] = 0.1*i + y[i]
print(y)
plt.plot(range(x),y)
plt.show()
```

## Find local min/max

```python
loc_max = argrelextrema(y, np.greater)
loc_min = argrelextrema(y, np.less)
print(loc_max, type(loc_max))
print(loc_min, type(loc_min))
```

## Plot min/max points as vertical lines

```python
plt.figure(figsize=(20,10))
plt.plot(range(x), y, 'k')
for i in range(len(loc_max[0])):
    plt.axvline(x=loc_max[0][i], color='b', linewidth=1.5)
for i in range(len(loc_min[0])):
    plt.axvline(x=loc_min[0][i], color='r', linewidth=2)
plt.show()

#max in blue, min in red
```
